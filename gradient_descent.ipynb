# Using Own Batch Gradient Descent

```python
 class Batch_Grad:
     
     def __init__(self,learning_rate,loops):
         self.coef = None
         self.intercept = None
         self.learning_rate = learning_rate
         self.loops = loops
     def fit(self,x_train,y_train):
        self.intercept = 0
        self.coef = np.ones(x_train.shape[1])
        for i in range(self.loops):
            #updating intercept values
            y_hat = np.dot(x_train,self.coef)+self.intercept
            # print(y_hat.shape)
            intercept_der = -2*np.mean(y_train-y_hat)
            self.intercept = self.intercept-(self.learning_rate*intercept_der)
            #updating coefs values
            coef_der = -2 * np.dot((y_train-y_hat),x_train)/x_train.shape[0]
            self.coef = self.coef - (self.learning_rate*coef_der)
            
        print(self.intercept,self.coef)
     def predict(self,x_test):
        return np.dot(x_test,self.coef)+self.intercept
```

```python
batch = Batch_Grad(0.1,100)
```

```python
start = time.time()
batch.fit(x_train,y_train)
print("Time taken by algorithom",time.time()-start)
```

```python
y_preds = batch.predict(x_test)
```

```python
r2_score(y_test,y_preds)
```

# Stochastic Gradient Descent

```python
 class SGDD_Regressor:
     
     def __init__(self,learning_rate,loops):
         self.coef = None
         self.intercept = None
         self.learning_rate = learning_rate
         self.loops = loops
     def fit(self,x_train,y_train):
        self.intercept = 0
        self.coef = np.ones(x_train.shape[1])
        for i in range(self.loops):
            for j in range(x_train.shape[0]):
                #get a random value
                index = np.random.randint(0,x_train.shape[0])
                #updating intercept values
                y_hat = np.dot(x_train[index],self.coef)+self.intercept
                # print(y_hat.shape)
                intercept_der = -2*(y_train[index]-y_hat)
                self.intercept = self.intercept-(self.learning_rate*intercept_der)
                #updating coefs values
                coef_der = -2 * np.dot((y_train[index]-y_hat),x_train[index])
                self.coef = self.coef - (self.learning_rate*coef_der)
                
        print(self.intercept,self.coef)
     def predict(self,x_test):
        return np.dot(x_test,self.coef)+self.intercept
```

```python
sgd = SGDD_Regressor(0.1,100)
```

```python
start = time.time()
sgd.fit(x_train,y_train)
print("Time taken by algorithom",time.time()-start)
```

```python
y_preds = sgd.predict(x_test)
```

```python
r2_score(y_test,y_preds)
```
